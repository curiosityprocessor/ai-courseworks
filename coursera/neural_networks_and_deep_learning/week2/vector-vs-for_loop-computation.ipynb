# The computational efficiency for vectorization
Why vectorization is the efficient method versus manual for-loop.  

## Code example

```python
import numpy as np
import time

a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = np.dot(a, b)
toc = time.time()
print(f"Vectorized version execution time: {(toc - tic) * 1000} ms")

c = 0;
tic = time.time();
for i in range(1000000):
    c += a[i] * b[i]
toc = time.time()
print(f"For loop version execution time: {(toc - tic) * 1000} ms")

```

## Why vectorization is more efficient
Modern CPU/GPU allow SIMD (Single Instruction, Multiple Data) operations for parallel execution.  
Python library such as numpy is more optimized to utilize parallelism.
